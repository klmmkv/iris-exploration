<h1> 2. KNN </h1>

KNN is used to make simple predictions for a new sample of data. KNN starts by training on the dataset which is already there. It trains, by calculating the euclidean distances from the "new" points, to each "old" point. These distances are then stored and sorted from nearest to furthest. Then the value for "k" is decided in order to pic the nearest neighbors of the new datapoint(s). Let's say we take "3" as the value for "k", then the top 3 neighbors would be taken. From them we see to which dataset they belong or what result ist the most frequent one. So if two of the 3 neighbors are "setosa" then the prediction would also be "setosa" for the new datapoint(s). 

```python
#1 Split up the training set. 

#load the dataset
iris = datasets.load_iris()

#get it into a list shape
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])

#call it variable dataset
dataset = df.values.tolist()

dataset = np.array(dataset)
dataset


D = dataset

#2. Create an array of indices for the data
D_indices = np.arange(0,len(D),1)
random_indices = np.random.permutation(D_indices) 

#3.1 split the indices into 5 folds and assign a test set 'outerset test' and a train set 'outer train'

#1 Fold Outer

#Outer Set 
split_indices = np.array_split(random_indices, 5)
outer1test = split_indices[4]

selected = []
for idx in [0,1,2,3]:
    selected.append(split_indices[idx][:])

outer1train = np.concatenate(selected)
```

```python
"""

the k-nearest-neighbor algorithm is by itself simple. I would say it is devided up into 3 main steps. 
#1. Calculate euclidean distance and sort the values from nearest to furthest. 
#2. Decide a value for "k" and point out the most frequent one out of the k-neighbors.
#3. Make the Prediction 

Here I calculate the euclidean distance. 
My input values are the training set, the test set, and the lenght of the test set

"""

outer1test = D[outer1test,:]
outer1train = D[outer1train,:]

def euclidean_distance_sorted(trainset, testset, length):
    distanceslist = list()
    for testvalue in testset:
        for trainvalue in trainset:
            #First I am calculating the distance for each point in the test set against each point in the training set. 
            dist = np.linalg.norm(testvalue[0:4]-trainvalue[0:4])
            #Then I append the received distance and the testpoint to a list 'distances'
            distances = (dist, testvalue[0:5])
            distanceslist.append(distances)
            
    
    #Here i break it back down, so every test value is in a different list
    chunks = [distanceslist[x:x+length] for x in range(0, len(distanceslist), length)]
    #Here I sort each list of the test values in order to get the nearest distances on top
    #So I know which points are the closest to each "new" datapoint
    sorteddistances = [sorted(lst, key=lambda t: t[0]) for lst in chunks]
    return sorteddistances
    
sorteddistances = euclidean_distance_sorted(outer1train, outer1test, 90)

from collections import Counter 

#This is the part where the prediction for each "test"/"new" point is made
#The function takes in the list with the sorteddistances and the number of neighbors 
def prediction(database, num_nei):
    
    #First we cut the list to have only the amount of 'num_nei' neighbors.
    neighborslists = list()
    for rows in database:
        num = num_nei
        points = rows[0:num]
        neighborslists.append(points)

    #just get the top points
    predictedvalues = list()
    for lists in neighborslists:
        for rows in lists:
            values = rows[1]
            predictedvalues.append(values)
    
    #only get the predicted classes 
    values = list()
    for rows in predictedvalues:
        dist = rows[4]
        values.append(dist)


    prediction = [values[x:x+num_nei] for x in range(0, len(values), num_nei)]

    #In the end I count the most common value in the set in order to give the prediction
    topprediction = []
    for ele in prediction:
        topprediction.append(Counter(ele).most_common(1)[0][0])
    return topprediction

prediction(sorteddistances, 3)
```

<h1>3. CLASSIFIER</h1>

```python
"""

In order to check if our algorithm is perfoming well, we need to understand whith which accuracy it works. 
The process is split up in two parts, firstly, I calculate the accuracy, secondly, I create the confusion matrix 

I take the predictions and the testset as the input for the function "accurancy" as I want to compare them.
Then we take all the predictions and compare them to the real values of the test set. 
I calculate how many "matches" they had and devide it against the actual length of it to get a percentage. 
In the end a confusion matrix is created. 

"""

#Taking the predictions list with the number of neighbots
topprediction = prediction(sorteddistances, 3)

def accurancy(topprediction, testset):
    #I get only the classification values form the testset - (0,1,2) - and create a list "actual". 
    actual = list()
    for i in testset:
        values = i[4]
        actual.append(values)

    #here I look if the values of the predictions match the values of actual and store it in the variable "matches"
    matches = [i for i, j in zip(topprediction, actual) if i == j]
    #In the end I devide the 
    accuracy = sum(matches)/len(actual)
    

    y_actu = pd.Series(actual, name='Actual')

    y_pred = pd.Series(topprediction, name='Predicted')

    df_confusion = pd.crosstab(y_pred, y_actu)
    return accuracy, df_confusion
    
    
accurancy(topprediction, outer1test)
```

```python
"""

For Nested cross-validation the process was a little harder and unfortunately manual in the end,
as I had some difficulties to get the loop in the end. However for the not so big - 
Iris Dataset it was managable this way. 

K-Fold cross-validation is when you split up your dataset into K - partitions  5.
For each partition, a model is fitted to the current split of training and testing dataset. 

Nested-Cross Validation is an extension of the above. It from the training set, we create another set, and 
split it up once more into a validation and test set. So for each outer fold in our cross validation, we have
another inner fold of "nested" cross validation. In this nested fold we calculate the best hyperparameters, and 
then train the outerfold with the best parameters. In the end we train the outerfolds again an understand 
which is the best training data, best parameters etc so that it is the most stable for new test sets. 

Nested-Cross Validation helps to prevent information leakage or cause significant bias. 
So it is more unbiased and with low variance. It is mostly used on smaller datasets where it would not be 
as comutationally expensive as for huge datasets. 
"""

#load the dataset
iris = datasets.load_iris()

#get it into a list shape
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])

#call it variable dataset
dataset = df.values.tolist()

dataset = np.array(dataset)
dataset


D = dataset

#Create an array of indices for the data
D_indices = np.arange(0,len(D),1)
random_indices = np.random.permutation(D_indices) 

```

```python
#3.1 split the indices into 5 folds and assign a test set 'outerset test' and a train set 'outer train'

#1 Fold -- outertest1 = [4] outertrain1 = [0,1,2,3]
#2 Fold --outertest2 = [3] outertrain2 = [0,1,2,4]
#3 Fold --outertest3 = [2] outertrain2 = [0,1,3,4]
#4 Fold --outertest4 = [1] outertrain2 = [0,2,3,4]
#5 Fold --outertest5 = [0] outertrain2 = [1,2,3,4]


#1 Fold Outer

#Outer Set 
split_indices = np.array_split(random_indices, 5)
outer1test = split_indices[4]

selected = []
for idx in [0,1,2,3]:
    selected.append(split_indices[idx][:])

outer1train = np.concatenate(selected)
```

```python
#1 Fold Inner

#Inner Set split it up into 3 Training Parts and 1 Validation Part

inner1train = np.array_split(outer1train, 4)
inner1test = inner1train[3]

inner1test

selectedinner1 = []

for idx in [0,1,2]:
    selectedinner1.append(inner1train[idx][:])

inner1train = np.concatenate(selectedinner1)


#convert back to original data so we can work with it
inner1test = D[inner1test,:]
inner1train = D[inner1train,:]


#calculate the distances from the testset to the trainset
distanceslist1 = list()

for testvalue in inner1test:
    for testtrain in inner1train:
        dist = np.linalg.norm(testvalue[0:4]-testtrain[0:4])
        distances = (dist, testtrain[0:5])
        distanceslist1.append(distances)
        

#split into distances for every testpoint
distancechunks1 = [distanceslist1[x:x+90] for x in range(0, len(distanceslist1), 90)]

#sort distances from nearest to furthest
sorteddistances1 = [sorted(lst, key=lambda t: t[0]) for lst in distancechunks1]



#now checking the accuracy of the test sets


#the function accuracyk takes in 3 parameters, the sorted dataset, the number of neighbors and the testset

def accuracyk(database, num_nei, inner1test):
    
    #I get the points which correlate to the amount of neighbors
    neighborslists = list()
    for rows in database:
        num = num_nei
        points = rows[0:num]
        neighborslists.append(points)

    #I only take the point, without the distance
    predictedvalues = list()
    for lists in neighborslists:
        for rows in lists:
            values = rows[1]
            predictedvalues.append(values)

    #and get the value (0,1,2) for the classification
    values = list()
    for rows in predictedvalues:
        dist = rows[4]
        values.append(dist)

    #then I create a variable with the amount of points and the classification to the point (0,1,2)
    prediction = [values[x:x+num_nei] for x in range(0, len(values), num_nei)]

    #Here I count the top value in each of the lists and give out the most frequent value
    from collections import Counter 
    
    topprediction = []
    for ele in prediction:
        topprediction.append(Counter(ele).most_common(1)[0][0])

    #then I create the list with the actual classification data of the validation set in order to compare both
    actual = list()
    for i in inner1test:
        values = i[4]
        actual.append(values)

    #comparison and getting the accuracy of how many match
    matches = sum(1 for p, a in zip(topprediction, actual) if p == a)
    accuracy = matches/len(actual)
    print(accuracy)
    return accuracy
    

accuracyk(sorteddistances1, 1, inner1test)
accuracyk(sorteddistances1, 2, inner1test)
accuracyk(sorteddistances1, 3, inner1test)
accuracyk(sorteddistances1, 4, inner1test)
accuracyk(sorteddistances1, 5, inner1test)
accuracyk(sorteddistances1, 6, inner1test)
accuracyk(sorteddistances1, 7, inner1test)
accuracyk(sorteddistances1, 8, inner1test)
accuracyk(sorteddistances1, 9, inner1test)
accuracyk(sorteddistances1, 10, inner1test)
```
